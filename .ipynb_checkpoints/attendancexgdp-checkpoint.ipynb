```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('avg_audience_championship_brasileiro.csv')
df.head()
```

```python
df = df.dropna()
df['Publico'] = df['Publico'].replace('-', np.nan)
df['Ano'] = df['Ano'].astype(str).str[:4]
df = df[12:]


df['Ano'] = pd.to_datetime(df['Ano'])
df['Publico'] = df['Publico'].astype(int)
df.drop(df.index[-1], inplace=True)
df.columns = ['Year', 'Attendance']
df.head()
```

```python
gdp = pd.read_csv('API_NY.GDP.PCAP.KD.ZG_DS2_en_csv_v2_10473715.csv', skiprows=3)
gdp.head()
```

```python
bra1 = gdp[gdp['Country Name'] == 'Brazil']
bra1 = bra1.T
bra1 = bra1.reset_index()
bra1 = bra1[4:]
bra1.columns = ['Year', 'GDP']
bra1 = bra1.dropna()
bra1 = bra1[8:]
bra1.head()
```

```python
%matplotlib notebook
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(9,5))
ax1.plot(df['Year'], df['Attendance'])
ax1.set_title('Brazilian Soccer Championship Average Attendance 1969-2017')

ax2.plot(df['Year'], bra1['GDP'], color='orange')
ax2.set_title('Brazil GDP Per Capita 1969-2017')

ax1.set_ylabel('Attendance')
ax2.set_ylabel('Growth (% Annual)')

plt.tight_layout()

plt.show()
```

```python
fig.savefig('attendancexgdp2.png', format='png')
```

```python
%matplotlib notebook
fig, ax1 = plt.subplots(figsize=(9,5))

ax1.set_ylabel('Attendance')
ax1.set_title('Brazilian Soccer Championship Average Attendance\nx\nGrowth Annual Brazil GDP Per Capita (1969-2017)')
ax1.plot(df['Year'], df['Attendance'])



ax2 = ax1.twinx()

ax2.set_ylabel('Growth (% Annual)')
ax2.plot(df['Year'], bra1['GDP'], color='orange')
plt.tight_layout()

plt.legend((ax1.lines[0], ax2.lines[0]), ('Attendance', 'GDP'))


plt.show()
    
```
